```python
#extracting zip file dataset 
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()
zip_ref.close()
```

```python
#reading dataset
import pandas as pd
df=pd.read_csv('orders.csv',na_values=['Not Available', 'unknown']) # handling null values in dataset
df.sample(10)
```

```python
#making column names into lowercase and converting space into underscore for easy understanding
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
df
```

```python
#deriving discount amount and  sales amount and profit
df['discount']=df['discount_percent']*df['list_price']*.01
df['discount']
```

```python
df['sales_price']=df['list_price']-df['discount']
df['sales_price']
```

```python
df['profit']=df['sales_price']-df['cost_price']
df['profit']
```

```python
df
```

```python
df.dtypes
```

```python
#change order_date from object to date data type
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
df['order_date']
```

```python
#drop unnecessary columns
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
```
